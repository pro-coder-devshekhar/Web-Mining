### **2. Extraction of different components of the page**

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://en.wikipedia.org/wiki/Website')
bs = BeautifulSoup(html.read(), 'html.parser')
print(bs.h1)
```

```python
contents = bs.findAll('span', {'class': 'toctext'})
for content_heading in contents:
  print(content_heading.get_text())
```

```python
for link in bs.find_all('a'):
  if 'href' in link.attrs:
    print(link.attrs['href'])
```

```python
nameList = bs.find_all(text='Website')
print(len(nameList))
```

### **3.Error Handling**

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
try:
  html = urlopen("https://en.wikipedia.org/wiki/Websitee")
except HTTPError as e:
  print("The server returned an HTTP error")
except URLError as e:
  print("The server could not be found!")
else:
  print(html.read())
```

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

def getTitle(url):
  try:
    html = urlopen(url)
  except HTTPError as e:
    return None
  try:
    bsObj = BeautifulSoup(html.read(), "lxml")
    title = bsObj.body.h1
  except AttributeError as e:
    return None
  return title

title = getTitle("https://en.wikipedia.org/wiki/Website")
if title == None:
  print("Title could not be found")
else:
  print(title)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('https://en.wikipedia.org/wiki/Website')
bs = BeautifulSoup(html, 'html.parser')

for sibling in bs.find('table').tr.next_siblings:
  print(sibling)
```
